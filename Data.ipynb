```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from statistics import stdev 
```

```python
data = np.loadtxt('accel-run5.csv')
```

```python
fig,ax = plt.subplots(1)
ax.scatter(data[:,0], data[:,1]); #time v x
fig,ax = plt.subplots(1)
ax.scatter(data[:,0], data[:,2]); #time v y
fig,ax = plt.subplots(1)
ax.scatter(data[:,0], data[:,3]); #time v z
```

```python
xstd= stdev(data[:,1])
```

```python
N = 100
l = lambda d: int(len(d) / N)

stdx = np.zeros(N)
time = np.zeros(N)
for n in range(l(data), len(data), l(data)):
    #boundary of interval starting at the end of the previous interval 
    #for i in range(n - l(data), n):
    time[min(int(n/l(data)) - 1, len(time) - 1)] = data[:,0][int((n + (n - l(data)))/2)]
    stdx[min(int(n/l(data)) - 1, len(time) - 1)] = stdev(data[:,1][n - l(data):n])

stdx[9] = stdev(data[:,1][n:])
time[9] = data[:,0][int((n + len(data))/2)]
```

```python
fig,ax = plt.subplots(dpi=120)
ax.plot(time, stdx, '.'); 

ax.set_title(r'Standard Deviaton of $a_x$ over time')
plt.xlabel('time(s)')
plt.ylabel('std dev')
fig.savefig("ax_std.png", dpi=120)
```
